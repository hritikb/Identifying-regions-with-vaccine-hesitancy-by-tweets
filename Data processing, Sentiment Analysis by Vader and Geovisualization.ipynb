```python
# Function for data cleaning
def clean(text = '', stopwords = []):
    #tokenize
    tokens = word_tokenize(text.strip())      #Returns syllables (or vaguely words, numbers and punctuation marks) as a list.
    
    #lowercase
    clean = [i.lower() for i in tokens]
    
    #remove stopwords
    clean = [i for i in clean if i not in stopwords]  #Makes the list free from stopwords like this, is etc.
    
    #remove punctuation
    punctuations = list(string.punctuation)  #returns a list of punctuation
    clean = [i.strip(''.join(punctuations)) for i in clean if i not in punctuations]   #Makes the list free from punctuations
    
    return ' '.join(clean)
```

```python
df['clean_text'] = df['text'].apply(str)  #make sure description is a string.
df['clean_text'] = df['clean_text'].apply(lambda x: clean(text = x, stopwords = stopwords.words('english')))
```

```python
# function to mark only Covid-19 vaccine related tweets
def vaccine_tweet(x):
    for i in lst:
        if i in x:
            return True
    
    return False
```

```python
df['vaccine_related'] = df['clean_text'].apply(vaccine_tweet)
```

# Sentiment analysis of tweets using VADER

```python
import numpy as np
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv('October 2021 Data Ready for Sentiment Analysis.csv', index_col = 0)
```

```python
df.head()
```

```python
df.shape
```

```python
# Applying the vader tool to the cleaned tweets for getting their sentiment scores
df['score'] = df['clean_text'].apply(lambda x: sid.polarity_scores(x))
df.head()
```

```python
# the value of compound in the score column gives the overall sentiment of the tweet, so we make a separate column for that
df['compound'] = df['score'].apply(lambda x: x['compound'])
```

```python
df.head()
```

```python
# function to separate tweets as positive, negative or neutral
def overall_sentiment(comp):
    if comp > 0:
        return 'positive'
    elif comp < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
df['sentiment'] = df['compound'].apply(overall_sentiment)  # Applying the function overall_sentiment to compound
```

```python
df.head()
```

```python
# function to find latitude and longitude for place of a tweet
def x_coordinate(obj):
    try:
        
        obj = str(obj)
        i = obj.find('coordinates')
        obj = obj[i:]
        j = obj.find(',')

        return float(obj[15:j])
    except Exception as e:
        return np.nan

def y_coordinate(obj):
    try:
        i = obj.find('coordinates')
        obj = obj[i:]
        j = obj.find(',')
        k = obj.find(']')

        return float(obj[j+2:k])
    except Exception as e:
        return np.nan
```

```python
# applying the functions to get coordinates from 'place' 
df['x_coordinate'] = df['place'].apply(x_coordinate)
df['y_coordinate'] = df['place'].apply(y_coordinate)
```

```python
# removing those tweets for which coordinates couldn't be extracted
df = df[~df['x_coordinate'].isnull() == True]
```

```python
df.to_csv('October 2021 Data with Sentiments and Coordinates.csv')
```

# Geovisualization

```python
import pandas as pd #importing libraries
import folium
```

```python
df = pd.read_csv('October 2021 Data with Sentiments and Coordinates.csv') #reading the sentiment analysed csv
df.head()
```

```python
#creating a map that's centered to our sample
some_map = folium.Map(location=[df['y_coordinate'].mean(), 
                                df['x_coordinate'].mean()], 
                                zoom_start=4.9)

for row in df.itertuples():
    if row.sentiment == "positive":
        some_map.add_child(folium.CircleMarker(location=[row.y_coordinate, row.x_coordinate],
                                         radius=2,color = "green", fill_color='green'))
    elif row.sentiment == "negative":
        some_map.add_child(folium.CircleMarker(location=[row.y_coordinate, row.x_coordinate],
                                         radius=2,color = "red" ,fill_color='red'))

# We are not plotting the neutral tweets because they are not really adding much meaning to our analysis  

#     elif row.sentiment == "neutral":
#         some_map.add_child(folium.CircleMarker(location=[row.y_coordinate, row.x_coordinate],
#                                          radius=1,color = "blue", fill_color='blue'))

some_map
```
